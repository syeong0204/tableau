```python
import pandas as pd
```

```python
file_to_load = 'Resource\JC-202101-citibike-tripdata.csv'
```

```python
jan21 = pd.read_csv(file_to_load)
jan21
```

```python
jan21_df = jan21.rename(columns = {
    'starttime': 'started_at', 'stoptime': 'ended_at', 'start station name': 'start_station_name', 'end station name': 'end_station_name',
    'end station id': 'end_station_id', 'start station latitude': 'start_lat', 'start station longitude': 'start_lng',
    'end station id': 'end_station_id', 'end station name': 'end_station_name', 'end station lattitude': 'end_lat',
    'end station longitude': 'end_lng', 'usertype': 'member_casual'
})
jan21_df
```

```python
membs = jan21_df['member_casual']
```

```python
membs
```

```python
for name in range(len(membs)):
    if membs[name] == 'Subscriber':
        membs[name] = 'member'

        
```

```python
for name in range(len(membs)):
    if membs[name] == 'Customer':
        membs[name] = 'casual'
```

```python
membs
```

```python
jan21_df.to_csv(r'JC-202101-citibike-tripdata.csv')
```
